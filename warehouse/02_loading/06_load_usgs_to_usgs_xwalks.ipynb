```python
import teehr
from teehr.evaluation.spark_session_utils import create_spark_session
```

```python
spark = create_spark_session(
    aws_profile="default"
)
```

```python
ev = teehr.RemoteReadWriteEvaluation(spark=spark)
```

```python
usgs_xwalk_sdf = ev.sql("""
    SELECT id as primary_location_id, id as secondary_location_id FROM locations WHERE id LIKE 'usgs-%'
""")
```

```python
usgs_xwalk_sdf.show()
```

```python
ev.location_crosswalks.load_dataframe(usgs_xwalk_sdf)
```

# Apply Migrations
This workbook applies the migrations in the `teehr-hub/warehouse/migrations` directory.  You must clone the repo into TEEHR-HUB and have write permissions to the data warehouse to run this.

```python
from teehr.utilities.apply_migrations import evolve_catalog_schema
import teehr
from teehr.evaluation.spark_session_utils import create_spark_session
from pathlib import Path
```

```python
spark = create_spark_session(
    aws_profile="default",
)
```

```python
migrations_dir_path = Path(Path().cwd().parents[0], "migrations")
migrations_dir_path
```

```python
evolve_catalog_schema(
    spark=spark,
    migrations_dir_path=migrations_dir_path,
    target_catalog_name="iceberg",
    target_namespace_name="teehr"
)
```

```python
spark.stop()
```

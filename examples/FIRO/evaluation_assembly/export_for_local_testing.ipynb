```python
import os
from pathlib import Path
import shutil

import pandas as pd
import numpy as np
import functools

import teehr

import pyspark as ps
from pyspark.sql import functions as F

teehr.__version__
```

```python
test_eval_dir = Path(Path().cwd(), "FIRO_test_evaluation")

# access local evaluation
ev = teehr.LocalReadWriteEvaluation(
    dir_path=test_eval_dir,
)

ev.enable_logging()
```

```python
# define folder to export 
output_dir = Path(Path().cwd().parent, "local_development", "v7_tables_parquet")
if output_dir.exists():
    shutil.rmtree(output_dir)
output_dir.mkdir(parents=True, exist_ok=True)
```

```python
# locations (use geopandas)
gdf = ev.table('locations').to_geopandas()
outpath = Path(output_dir, 'locations.parquet')
gdf.to_parquet(outpath)

# location_crosswalks
table_name = 'location_crosswalks'
df = ev.table(f'{table_name}').to_pandas()
outpath = Path(output_dir, f'{table_name}.parquet')
df.to_parquet(outpath)

# variables
table_name = 'variables'
df = ev.table(f'{table_name}').to_pandas()
outpath = Path(output_dir, f'{table_name}.parquet')
df.to_parquet(outpath)

# units
table_name = 'units'
df = ev.table(f'{table_name}').to_pandas()
outpath = Path(output_dir, f'{table_name}.parquet')
df.to_parquet(outpath)

# configurations
table_name = 'configurations'
df = ev.table(f'{table_name}').to_pandas()
outpath = Path(output_dir, f'{table_name}.parquet')
df.to_parquet(outpath)

# primary_timeseries
table_name = 'primary_timeseries'
df = ev.table(f'{table_name}').to_pandas()
outpath = Path(output_dir, f'{table_name}.parquet')
df.to_parquet(outpath)

# secondary_timeseries
table_name = 'secondary_timeseries'
df = ev.table(f'{table_name}').to_pandas()
outpath = Path(output_dir, f'{table_name}.parquet')
df.to_parquet(outpath)

# joined_timeseries
table_name = 'joined_timeseries'
df = ev.table(f'{table_name}').to_pandas()
outpath = Path(output_dir, f'{table_name}.parquet')
df.to_parquet(outpath)

# locations_metrics
table_name = 'locations_metrics'
df = ev.table(f'{table_name}').to_pandas()
outpath = Path(output_dir, f'{table_name}.parquet')
df.to_parquet(outpath)

# event_rankings
table_name = 'event_rankings'
df = ev.table(f'{table_name}').to_pandas()
outpath = Path(output_dir, f'{table_name}.parquet')
df.to_parquet(outpath)

# event_heatmap
table_name = 'event_heatmap'
df = ev.table(f'{table_name}').to_pandas()
outpath = Path(output_dir, f'{table_name}.parquet')
df.to_parquet(outpath)

# attributes
table_name = 'attributes'
df = ev.table(f'{table_name}').to_pandas()
outpath = Path(output_dir, f'{table_name}.parquet')
df.to_parquet(outpath)

# location_attributes
table_name = 'location_attributes'
df = ev.table(f'{table_name}').to_pandas()
outpath = Path(output_dir, f'{table_name}.parquet')
df.to_parquet(outpath)
```

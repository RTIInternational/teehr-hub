```python
from teehr import RemoteReadWriteEvaluation
from teehr.evaluation.spark_session_utils import create_spark_session
```

#### Start the spark session

```python
spark = create_spark_session(
    aws_access_key_id="minioadmin",
    aws_secret_access_key="minioadmin123",
    update_configs={
        "spark.hadoop.fs.s3a.aws.credentials.provider":
        "org.apache.hadoop.fs.s3a.AnonymousAWSCredentialsProvider"
    }
)
```

#### Initialize a TEEHR Evaluation with read/write privileges

```python
ev = RemoteReadWriteEvaluation(spark=spark)
```

#### Take a look at the configurations

```python
ev.configurations.to_sdf().show()
```

#### Create and write the joined timeseries table (pairing the primary and secondary timeseries)

```python
joined_timeseries = ev.joined_timeseries_view(
    secondary_filters=["configuration_name = 'nwm30_retrospective'"]
).write(table_name="sim_joined_timeseries")
```

```python
joined_timeseries.to_sdf().show()
```

```python
spark.stop()
```

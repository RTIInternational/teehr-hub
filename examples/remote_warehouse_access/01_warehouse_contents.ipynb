```python
import os
from pathlib import Path
import tempfile
import shutil

import teehr

teehr.__version__
```

```python
from teehr.evaluation.spark_session_utils import create_spark_session
spark = create_spark_session()
```

```python
ev = teehr.RemoteReadOnlyEvaluation(spark=spark)
```

```python
# List out tables in the TEEHR Warehouse
ev.list_tables()
```

```python
sdf = ev.configurations.to_sdf()
print(f"Count: {sdf.count()}")
sdf.show(10, False)
```

```python
sdf = ev.attributes.to_sdf()
print(f"Count: {sdf.count()}")
sdf.show(5, False)
```

```python
sdf = ev.variables.to_sdf()
print(f"Count: {sdf.count()}")
sdf.show(5, False)
```

```python
sdf = ev.units.to_sdf()
print(f"Count: {sdf.count()}")
sdf.show(5, False)
```

```python
sdf = ev.locations.filter("id like 'usgs-%'").to_sdf()
print(f"Count: {sdf.count()}")
sdf.show(5, False)
```

```python
sdf = ev.location_attributes.to_sdf()
print(f"Count: {sdf.count()}")
sdf.show(5, False)
```

```python
sdf = ev.location_crosswalks.to_sdf()
print(f"Count: {sdf.count()}")
sdf.show(5, False)
```

```python
sdf = ev.primary_timeseries.to_sdf()
print(f"Count: {sdf.count():,}")
sdf.show(5, False)
```

```python
sdf = ev.secondary_timeseries.to_sdf()
print(f"Count: {sdf.count():,}")
sdf.show(5, False)
```
